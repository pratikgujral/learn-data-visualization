## Introduction to relational plots and subplots
Questions about quantitative variables

---
To compare and find the relation between different quantitative variables, we used scatterplots. In order to add a third vairable, we used the **`hue`** parameter. However, we can also plot multiple scatterplots comparing the effect of the thrid variable. This can be done by plotting a **`relplot()`** (relational plot).

### `relplots()`
- Create relationship plots- usign scatterplots or lineplots
- An advantage of using `relplot()` is its ability to create subplots within the same figure



````
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
````

````
tips = sns.load_dataset('tips')
tips.head()
````

#### Scatterplot vs Relplot

````
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
````

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
````

We can add addtional vairbale through a subplot using the **`col`** parameter.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter',
            col='smoker')

plt.show()
````

If we want to arrange the two plots in two different rows instead of columns, we can use the **`row`** parameter instead.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter',
            row='smoker')

plt.show()
````

It is possible to use both- **`col`** and **`row`** at the same time.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter',
            col='smoker', row='time')

plt.show()
````

Another example...  

Here, we plot the `total_bill` vs `tip`, but with `day` (representing day of the week) as the subgroup parameter.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter',
            col='day')
plt.show()
````

So, as there are many different categories available in the `day` column, we are seeing multiple plots along the row. It can become tedious as the number of categories increase. So, to address this, we can use the **`col_wrap`** parameter. 

Using the **col_wrap** parameter, we specify how many plots per row we wish to see.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter',
            col='day', col_wrap=2) # Only 2 plots per row

plt.show()
````

If the plots are supposed to be plotted in a particular order such as- 'Thu', 'Fri', 'Sat', 'Sun' in the above example, we can do so by explicitly passing the **`col_order`** and **`row_order`** parameters.

````
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='day', col_wrap=2, col_order=['Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
````

## Introduction to line plots

**Line plots**
- Each point usually means the same "thing", typically used to track it over time

### Changing the whiskers using `whis`
- By default, the whiskers extend to  1.5 * IQR, where IQR is the range from 25th to 75th percentile of the distribution of data

- If we simply want the whiskers to appear at the min and max values, we may specify **`whis=[0, 100]`**

````
sns.catplot(x='time', y='total_bill', data=tips, kind='box',
            whis=[0, 100])
plt.show()
````

# Part-4: Customizing Seaborn plots

## Changing plot style and color
While Seaborn plot defaults are aesthetically pleasing, there are several reasons one would want to customize the appearance of the plots.
- Personal preference
- Improve readability
- Help orient an audeince more quickly to the key takeaway that you wish to dipic from the plot

### Preset styles
Seaborn has **5** preset styles that change the background and axis of the plots
- **`white`** (default)
- **`dark`**
- **`whitegrid`**
- **`darkgrid`**
- **`ticks`**

To set one of these as the global style for all of the plots, use **`sns.set_style('<style name>')`**

### Changing color palette
- Figure palette changes the color of the main elements of the plot
- **`sns.set_palette()`**
- There are several preset palettes available in Seaborn. 
  - The **Diverging palettes** are great to use if the visualization delas with a scale where two ends of the scale are opposites and there is a neutral midpoint. Eg. "RdBu", "PRGn", "RdBu_r", "PRGn_r"

  - **Sequential Palettes** are good for emphasing a variable on a continuous scale. Eg. "Greys", "Blues", "PuRd", "GnBu"

- We can also create **custom palettes** by passing in a list of color names or hex color codes.
```
custom_palette = ['red', 'green', 'orange', 'blue', 'yellow', 'purple']
sns.set_palette(custom_palette)
```

[Link to Seaborn documentation on palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)

### Changing the scale of the plot
- Scale of the plot can be changed using the **`sns.set_context()`** function
- Scale options smallest to largest are: `"paper"`, `"notebook"`, `"talk"`, and `"poster"`. Default context is `"paper"`
- Choose a larger scale like `"talk"`for posters or presentations if your audience is further away from the plot.

## Adding titles and labels
- Titles and axis labels are important to any plot

### `FacetGrid` vs `AxesSubplot` objects
- Seaborn plots create two different types of objects: **`FacetGrid`** and **`AxesSubplot`**
- To figure out which type of object we are working with, 

```
g = sns.scatterplot(x='height', y='weight', data=df)
type(g)
```

- An empty **`FacetGrid`** consists of one or more subplots, which is how it supports subplots. **`relplot()`** and **`catplot()`** both support creating subplots

<table>
<tr><th>Object Type</th><th>Plot Types</th><th>Characteristics</th></tr>
<tr><td><code>FacetPlot</code></td><td><code>relplot()</code>, <code>catplot()</code></td><td>Can create subplots</td></tr>
<tr><td><code>AxesSubplot</code></td><td><code>scatterplot()</code>, <code>countplot()</code> etc.</td><td>Only creates a single plot</td></tr>
</table>  

**The method for adding a title to your plot will differ depending on the type of object it is.**

### Adding a title to `FacetGrid`
Use **`g.fig.suptitle("<title>")`**

````
gdp_data = pd.read_csv('countries-of-the-world.csv')

gdp_data.head()
````

Dataset has comma instead of period to represent a decimal point

````
cols_numeric = ['Population', 'Area (sq. mi.)',
       'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
       'Net migration', 'Infant mortality (per 1000 births)',
       'GDP ($ per capita)', 'Literacy (%)', 'Phones (per 1000)', 'Arable (%)',
       'Crops (%)', 'Other (%)', 'Climate', 'Birthrate', 'Deathrate',
       'Agriculture', 'Industry', 'Service']

for col_numeric in cols_numeric:
  if gdp_data[col_numeric].dtype == 'object':
    gdp_data[col_numeric] = gdp_data[col_numeric].str.replace(',', '.')
    gdp_data[col_numeric] = pd.to_numeric(gdp_data[col_numeric], errors='coerce')
````

````
sns.reset_defaults()

g = sns.catplot(x='Region', y='Birthrate', data=gdp_data, kind='box')

# Adding plot title
g.fig.suptitle("New title") # this tells Seaborn we need to set the title to the whole figure

plt.show()
````

To adjust the height of the title, we can use the **`y=`** parameter

````
g = sns.catplot(x='Region', y='Birthrate', data=gdp_data, kind='box')

# Adding plot title
g.fig.suptitle("New title", y=1.03) # this tells Seaborn we need to set the title to the whole figure

plt.show()
````

## Rotating x-tick labels
In this exercise, we'll continue looking at the miles per gallon dataset.
