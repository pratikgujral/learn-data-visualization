## Customizing data appearance
We can customize the appearance of data in our plots, while adding the data to the plot, using key-word arguments to the `plot` command.

We will now customize the appearance of the markers, the linestyle that is used, and the **color of the lines** and **markers** for our data.

### Color of lines, marker styles, Line style
**`color`**, **`marker`**, and **`linestyle`** keyword arguments are used for this purpose. 

```
# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
# Making plot line as blue and marking points with a dot
# Makign line stayle as dashed
ax.plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-NORMAL'], color='b',
        marker='o', linestyle='--')

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
# Making line in red, and marking points with downward 'v'
# Making line style as dashed
ax.plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color='r',
        marker='v', linestyle='--')

# Call the show function
plt.show()
```

### Customizing axis labels and adding titles
Customizing the axis labels requires using the `set_xlabel` and `set_ylabel` methods of the Axes object. Adding a title uses the `set_title` method.

In this exercise, we will customize the content of the axis labels and add a title to a plot.

```
fig, ax = plt.subplots()

ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# Customize the x-axis label
ax.set_xlabel('Time (months)')

# Customize the y-axis label
ax.set_ylabel('Precipitation (inches)')

# Add the title
ax.set_title('Weather patterns in Austin and Seattle')

# Display the figure
plt.show()
```

## Plotting time-series with different variables
[VIDEO](https://videos.datacamp.com/transcoded/13706_introduction_to_matplotlib_in_python/v2/hls-13706_ch2_2.master.m3u8)

- When we plot multiple time-series in a plto such as CO2 levels and temperature on the y-axis vs time on the x-axis, our plots may not look informative if the multiple y-axis plot values are on different scales.
- If the data for multiple plots is on a different scale it is better to use a secondary axis. This can be done using a **twin axis**- **`ax.twinx()`**.
- **ax.twinx()** means that the two plots share the same x-axis, but different y-axis
- **`ax.ticks_params()`** can be further used to change the tick colors

```
import matplotlib.pyplot as plt

# Initalize a Figure and Axes
fig, ax = plt.subplots()

# Plot the CO2 variable in blue
ax.plot(climate_data.index, climate_data['co2'], color='b')

# Create a twin Axes that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature in red
ax2.plot(climate_data.index, climate_data['relative_temp'], color='r')

plt.show()
```

**NOTE:** We were able to create the secondary y-axis using the `twinx()` which allowed us to create a shared-x axis for the two plots

### Defining a function that plots time-series data
Once we realize that a particular section of code that we have written is useful, it is a good idea to define a function that saves that section of code for us, rather than copying it to other parts of your program where we would like to use this code.

```
# Define a function called plot_timeseries that takes as input an Axes object (axes), data (x,y), a string with the name of a color and strings for x- and y-axis labels
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label with provided color
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

```
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_data.index, climate_data['co2'], "blue", "Time (years)", "CO2 levels")

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_data.index, climate_data['relative_temp'], "red", "Time (years)", "Relative temperature (Celsius)")

plt.show()
```

## Annotating time-series data
[VIDEO](https://videos.datacamp.com/transcoded/13706_introduction_to_matplotlib_in_python/v2/hls-13706_ch2_3.master.m3u8)
- When plots are built, sometimes it may be difficult to figure out exactly what we are trying to convey through the plot
- One way is to mark an arrow at a specific data point we are trying to highlight and add text next to the arrow
> `ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2015-10-06'), -0.2), arrowprops={})`
- `xy=` parameter is a tuple used to denote the data point that is to be highlighted
- `xytext=` parameter is a tuple used to denote the location of the text on the plot
- `arrowprops` (arrow properties) is a dictionary used to draw an arrow from the text pointing to the highlighted data point.
  - If dictionary is empty, an arrow with default properties is made
  - Arrow can be customized such as `arrowprops={"arrowstyle": "->", "color": "gray"}`


### Annotating a plot of time-series data
Annotating a plot allows us to highlight interesting information in the plot. For example, in describing the climate change dataset, we might want to point to the date at which the relative temperature first exceeded 1 degree Celsius.


```
fig, ax = plt.subplots()

# Plot the relative temperature data
ax.plot(climate_data.index, climate_data['relative_temp'])

# Annotate the date at which temperatures exceeded 1 degree
ax.annotate('> 1degree', (pd.Timestamp('2015-10-06'), 1))

plt.show()
```

```
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_data.index, climate_data['co2'], 'blue', 'Time (years)', 'CO2 levels')

# Create an Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_data.index, climate_data['relative_temp'], 'red', 'Time (years)', 'Relative temp (Celsius)')

# Annotate point with relative temperature >1 degree
ax2.annotate(">1 degree", xy=(pd.Timestamp('2015-10-06'), 1), xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={'arrowstyle': '->', 'color': 'gray'})

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

medals = pd.read_csv('medals_by_country.csv', index_col=0)
medals.head()
```

```
fig, ax = plt.subplots()

# Making a bar chart
ax.bar(medals.index, medals['Gold'])
plt.show()
```

This bar has `country` on the x-axis, and the height of each bar represents the number in that row.

As we can see, the labels on the x-axis are rather long. We can make them more legible by rotating the text so that the text can fit in the available space.

```
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of medals')

plt.show()
```

The bar chart not only shows us visually about the number of gold meads won by individual country, but also helps us make a comparison among them by the height of each bar.

We can add silver and broze medals as well to the same chart, but it is better to represent them on a **stacked bar chart**

```
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold']) # Bottom of the silver bars should be at the height of the previous call's data

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of medals')

plt.show()
```

```
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold']) # Bottom of the silver bars should be at the height of the previous call's data
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'])

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of medals')

plt.show()
```

To make this figure easier to read and understand, we add a **legend**, so that we know which color belongs to which medal

```
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'], label='Gold')
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'], label='Silver') # Bottom of the silver bars should be at the height of the previous call's data
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'],
       label='Bronze')

ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of medals')

# Add a legend
ax.legend()

plt.show()
```

## Quantitative comparisons: scatter plots
- Scatter plots are useful for bivariate comparisons


```
import pandas as pd
import matplotlib.pyplot as plt

climate_data = pd.read_csv('climate_data.csv', parse_dates=['date'], 
                           index_col='date')

fig, ax = plt.subplots()

# Creating the scatter plot
ax.scatter(climate_data['co2'], climate_data['relative_temp'])

# Setting axis labels
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative Temperature (Celsius)')

plt.show()
```

### Customizing scatter plots

```
eighties = climate_data['1980-01-01':'1989-12-31']
nineties = climate_data['1990-01-01':'1999-12-31']

fig, ax = plt.subplots()

ax.scatter(eighties['co2'], eighties['relative_temp'], 
           color='red', label='eighties')
ax.scatter(nineties['co2'], nineties['relative_temp'], 
           color='blue', label='nineties')


# Adding the legend
ax.legend()

# Setting axis labels
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative Temperature (Celsius)')

plt.show()
```

Encoding the third variable in the scatter plot through color

```
fig, ax = plt.subplots()

# Creating the scatter plot
ax.scatter(climate_data['co2'], climate_data['relative_temp'], 
           c=climate_data.index)

# Setting axis labels
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative Temperature (Celsius)')

plt.show()
```

Time information (which is contained in the index of our dataframe) is now also encoded as colors of the data points on the scatter plot. Older dates are encoded as blue, while the newer dates are encoded as yellow.

## Automating figures from data
VIDEO

### Why automate?
We can easily write functions that can automatically adjust according to the data
Advantages of automating
- Easy and speed
- Flexibility
- Robustness
- Reproducibility

```
import pandas as pd
import matplotlib.pyplot as plt

medalists = pd.read_csv('medalist_weights.csv')
medalists.head()
```

Let's day we don't know how many types of sports are available. Also, new sports may be added to the dataset later on.

```
sports = medalists['Sport'].unique()
print(sports)
```

Bar charts of heights of each of the sports

```
fig, ax = plt.subplots()

for sport in sports:
  sport_df = medalists[medalists['Sport']==sport]

  ax.bar(sport, sport_df['Height'].mean(), yerr=sport_df['Height'].std())

ax.set_ylabel('Height (cm)')
ax.set_xticklabels(sports, rotation=90)
plt.show()
```
